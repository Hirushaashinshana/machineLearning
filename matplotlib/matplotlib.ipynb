```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
x2 = np.arange(0,4.5,0.5)
# plt.figure(figsize=(4,2),dpi=200) # Resize the graph
plt.title('Plot',fontdict={
    'fontname':'Comic Sans MS',
    'fontsize':20})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot(x,y,label='2x',
             color='green',
             linewidth=2 ,
             marker='*',
             markersize=5,
             linestyle='--')
plt.plot(x2,x2**2,
            label='x^2',
            color='red',
            linestyle='-',
            marker='*',
            mfc='green',
            markersize='5')
plt.xticks(x)
plt.yticks(y)
plt.legend()

plt.savefig('mychart.png',dpi=300)
plt.show()
```

```python
bars = plt.bar(x,y)
bars[1].set_hatch('/')
plt.show()
```

```python
ind_inflation = pd.read_csv('/ML Datasets/India_GDP_Data.csv')
Years = ind_inflation['Year']
GDP_In_Billion_USD = ind_inflation['GDP_In_Billion_USD']
list = []
for year in Years :
    list.append(year)
list.sort()
print(list)
```

```python
GDP_list = []
for each_year_gdp in GDP_In_Billion_USD:
    GDP_list.append(each_year_gdp)
GDP_list.reverse()
print(GDP_list)
```

```python
plt.figure(figsize=(6,4),dpi=100)
plt.plot(list,GDP_list,label='GDP',color='blue')
plt.title('GDP Growth Of India 1961-2021')
plt.xlabel('Year')
plt.ylabel('GDP (in Billon USD)')
plt.legend()
plt.savefig('India_GDP_Growth.png',dpi=300)
plt.show()
```
